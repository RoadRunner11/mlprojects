```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import missingno as msno
import seaborn as sn

data =pd.read_csv('C:\\Users\\E\\Downloads\\[Ttorial.Com] Real-World Machine Learning Projects with Scikit-Learn [Video]\\winequality-white.csv', sep= ';')
data
# missing data detection
msno.matrix(data, figsize= (10, 3))

#Distribution
fig, axes = plt.subplots(nrows=2, ncols=1)
fig.set_size_inches(10, 20)
sn.boxplot(data=data, orient='v', ax=axes[0])
sn.boxplot(data= data, y='quality', orient='pH', ax=axes[1])

#correlation analysis
corrMat = data.corr()
mask = np.array(corrMat)
mask[np.tril_indices_from(mask)]= False
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
sn.heatmap(corrMat, mask= mask, vmax=.8, square= True, annot=True)
```
